```python
#No. 2 from the assingment
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        # Create a list of numbers from 1 to n
        numbers = list(range(1, n+1))

        # Multiply all the numbers in the list together
        product = 1
        for i in numbers:
            product = product *i

        return product

print(factorial(5))
print(factorial(10))
print(factorial(20))
print(factorial(50))

```

```python
#No. 3 from the assingment

def special_factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        # Create a list of numbers from 1 to n
        numbers = list(range(1, n+1))

        if n >= 4:
            numbers.remove(4)
        if n >= 13:
            numbers.remove(13)
        if n >= 22:
            numbers.remove(22)

        # Multiply all the numbers in the modified list together
        product = 1
        for num in numbers:
            product *= num

        return product

# Test the special_factorial function
print(special_factorial(5))
print(special_factorial(10))
print(special_factorial(20))
print(special_factorial(50))

```
